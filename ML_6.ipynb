````python
# DA-AG-009: EDA Assignment – Bike Details Dataset

## Question 1: Read the Bike Details dataset into a Pandas DataFrame and display its first 10 rows.

### Python Code

```python
import pandas as pd

# Read dataset
df = pd.read_csv('Bike_Details.csv')

# Display first 10 rows
print(df.head(10))

# Display shape
print("Shape of dataset:", df.shape)

# Display column names
print("Column names:")
print(df.columns)
```

### Answer

The dataset was successfully loaded into a Pandas DataFrame. The first 10 rows were displayed using `head(10)`. The shape of the dataset and the column names were also examined to understand the structure of the data.

---

## Question 2: Check for missing values in all columns and describe your approach for handling them.

### Python Code

```python
# Check missing values
print(df.isnull().sum())

# Percentage of missing values
print((df.isnull().sum()/len(df))*100)
```

### Answer

The missing values were identified using the `isnull().sum()` function.

**Approach for handling missing values:**

* For numerical columns, missing values can be replaced using mean or median values.
* For categorical columns, missing values can be replaced using the mode.
* If a column contains a very large percentage of missing values, it may be removed from the dataset.

Example:

```python
df['km_driven'].fillna(df['km_driven'].median(), inplace=True)
df['owner'].fillna(df['owner'].mode()[0], inplace=True)
```

---

## Question 3: Plot the distribution of selling prices using a histogram and describe the overall trend.

### Python Code

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df['selling_price'], bins=30)
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.title('Distribution of Selling Prices')
plt.show()
```

### Answer

The histogram shows that the distribution of selling prices is positively skewed. Most bikes have relatively lower selling prices, while a few premium bikes have very high prices, creating a long right tail.

---

## Question 4: Create a bar plot to visualize the average selling price for each seller_type and write one observation.

### Python Code

```python
avg_price = df.groupby('seller_type')['selling_price'].mean()

avg_price.plot(kind='bar')
plt.xlabel('Seller Type')
plt.ylabel('Average Selling Price')
plt.title('Average Selling Price by Seller Type')
plt.show()

print(avg_price)
```

### Answer

**Observation:** Dealer-listed bikes generally have a higher average selling price compared to bikes sold by individual owners.

---

## Question 5: Compute the average km_driven for each ownership type and present the result as a bar plot.

### Python Code

```python
avg_km = df.groupby('owner')['km_driven'].mean()

avg_km.plot(kind='bar')
plt.xlabel('Owner Type')
plt.ylabel('Average KM Driven')
plt.title('Average KM Driven by Ownership Type')
plt.show()

print(avg_km)
```

### Answer

The analysis indicates that bikes with multiple owners tend to have higher average kilometers driven than first-owner bikes.

---

## Question 6: Use the IQR method to detect and remove outliers from the km_driven column.

### Python Code

```python
# Before removing outliers
print("Before:")
print(df['km_driven'].describe())

Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df_clean = df[(df['km_driven'] >= lower) &
              (df['km_driven'] <= upper)]

print("\nAfter:")
print(df_clean['km_driven'].describe())
```

### Answer

The IQR method successfully identified extreme values in the `km_driven` column. Removing these outliers reduced the effect of unusually high mileage values and improved the overall data quality.

---

## Question 7: Create a scatter plot of year vs. selling_price.

### Python Code

```python
plt.figure(figsize=(8,5))
plt.scatter(df['year'], df['selling_price'])
plt.xlabel('Year')
plt.ylabel('Selling Price')
plt.title('Year vs Selling Price')
plt.show()
```

### Answer

The scatter plot shows a positive relationship between manufacturing year and selling price. Newer bikes generally have higher resale values compared to older bikes.

---

## Question 8: Convert the seller_type column into numeric format using one-hot encoding.

### Python Code

```python
encoded_df = pd.get_dummies(df, columns=['seller_type'])

print(encoded_df.head())
```

### Answer

One-hot encoding converted the categorical `seller_type` column into multiple binary columns, making the dataset suitable for machine learning algorithms.

---

## Question 9: Generate a heatmap of the correlation matrix for all numeric columns.

### Python Code

```python
import seaborn as sns

corr = df.corr(numeric_only=True)

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### Answer

The correlation heatmap indicates that:

* The `year` variable usually has a positive correlation with `selling_price`.
* The `km_driven` variable often shows a negative correlation with `selling_price`.
* Strong positive and negative correlations highlight the most influential factors affecting bike prices.

---

## Question 10: Summarize your findings in a brief report.

### Answer

### Important factors affecting bike selling price:

1. Manufacturing year significantly affects selling price.
2. Kilometers driven negatively impact resale value.
3. Seller type influences the average selling price.
4. Ownership history affects bike valuation.

### Data cleaning performed:

* Checked and handled missing values.
* Removed outliers using the IQR method.
* Converted categorical variables using one-hot encoding.

### Feature engineering performed:

* One-hot encoding of categorical features.
* Creation of a cleaned dataset after outlier removal.

### Overall conclusion:

The exploratory data analysis revealed that newer bikes with fewer kilometers driven and fewer previous owners tend to have higher selling prices. Data cleaning and feature engineering improved the quality of the dataset and prepared it for future machine learning tasks such as price prediction.
````
